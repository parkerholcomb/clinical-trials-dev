```python
import plotly.express as px
import pandas as pd

from openctgov.lib.helpers import protocol_feature_query

protocol_feature = protocol_feature_query()
protocol_feature = protocol_feature[protocol_feature['_start_yr'] > 1999]
protocol_feature = protocol_feature[protocol_feature['_start_yr'] < 2023]
px.box(
    protocol_feature,
    x="_start_yr",
    y="_arm_count",
    title=f"n={len(protocol_feature)} Studies",
    facet_col="_random",
)
```

```python
# ok what about median per year
# protocol_feature

dfg = protocol_feature.groupby(['_start_yr']).agg(
    arm_count_mean = ("_arm_count","mean"),
    arm_count_median = ("_arm_count","median")
).reset_index()
px.scatter(dfg, x="_start_yr", y='arm_count_mean', trendline='ols')
```
