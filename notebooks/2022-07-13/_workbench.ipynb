```python
from src.ctgov.lib.helpers import protocol_feature_query

from nltk import ngrams, RegexpTokenizer

def _to_tokens(txt: str) -> list:
    tokenizer = RegexpTokenizer(r"\w+") 
    tokens = tokenizer.tokenize(txt)
    lower_tokens = [t.lower() for t in tokens]
    return lower_tokens

protocol_feature = protocol_feature_query()
protocol_feature
```

```python
import pandas as pd
pd.options.display.max_columns = None


df = protocol_feature #[['_condition']].reset_index(drop=True)
df["_condition_count"] = df["_condition"].apply(lambda x: len(x))
df = df[df["_condition_count"]==1].reset_index(drop=True)
df["_cx"] = df["_condition"].apply(lambda x: x[0])
df['_tokens'] = df['_cx'].apply(_to_tokens)

match_phrase = "addiction"
df['_match'] = df['_tokens'].apply(lambda x: match_phrase in x)
df = df[df['_match']].reset_index(drop=True)
df
```

```python
import plotly.express as px

px.histogram(df, x='_condition_count')
```
